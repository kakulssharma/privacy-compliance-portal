```python
import tenseal as ts

def create_context():
    ctx = ts.context(
        ts.SCHEME_TYPE.BFV,
        poly_modulus_degree=8192,
        plain_modulus=1032193
    )
    ctx.generate_galois_keys()
    return ctx

context = create_context()
```

```python
def check_eligibility(balance, threshold=15000):
    enc_balance = ts.bfv_vector(context, [balance])
    enc_threshold = ts.bfv_vector(context, [threshold])

    enc_diff = enc_balance - enc_threshold
    diff = enc_diff.decrypt()[0]

    return diff
```

```python
import json
import os

def write_zkp_input(diff):
    data = {
        "diff": diff
    }

    with open("zkp/input.json", "w") as f:
        json.dump(data, f)
```

```python
balance = int(input("Enter balance: "))
threshold = 15000

diff = check_eligibility(balance, threshold)
write_zkp_input(diff)

print("BFV diff:", diff)
print("ZKP input generated")
```
